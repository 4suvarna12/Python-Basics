```python
# 6 & 7 both are same question Write a function to demonstrate the behavior of mutable and immutable arguments
def demonstrate_mutability(mutable_arg, immutable_arg):


    print("Inside the function:")
    print("Mutable argument before modification:", mutable_arg)
    mutable_arg.append(4)  # Modifying the mutable argument
    print("Mutable argument after modification:", mutable_arg)

    print("Immutable argument before modification:", immutable_arg)
    immutable_arg += 1  # Modifying the immutable argument (creates a new object)
    print("Immutable argument after modification:", immutable_arg)


# Example usage
my_list = [1, 2, 3]
my_number = 10

demonstrate_mutability(my_list, my_number)

print("\nOutside the function:")
print("Mutable argument:", my_list)  # The original list is modified
print("Immutable argument:", my_number)  # The original number remains unchanged
```

```python
# 14.Write a program to reverse a string using a while loop
def reverse_string(s):
    reversed_string = ""
    i = len(s) - 1
    while i >= 0:
        reversed_string += s[i]
        i -= 1
    return reversed_string

input_string = "suvarna"
reversed_string = reverse_string(input_string)
reversed_string
```

```python
# 15.Write a Python program to calculate the factorial of a number provided by the user using a while loop
def fact(n):
  num=1
  while  n>=1:
    num=num *n
    n=n-1
  return num
f=fact(int(input("enter number")))
print(f)
```
